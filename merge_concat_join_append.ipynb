```python
import pandas as pd
```

```python
a = {
     'user_id':[22482,29628,28476,15200,28239],
     'marketing_name1':['ios','android','android', 'ios', 'android'],
    }
df = pd.DataFrame(a)
```

```python
b = {
    'marketing_name': ['sunny','bala','tejash','reshab','OX1020A'],
    'user_id': [22482,29628,22784,22488,28476]
    }
df2 = pd.DataFrame(b)
```
