```python
df.drop(index = 10472, inplace = True)
```

```python
df[10470:10475]
```

```python
percent = len(df) * 10/100
```

```python
percent
```

```python
df.dropna(thresh=percent, axis=1, inplace =True)
```

```python
df.isnull().sum()
```

```python
rating_fillvalue = df.Rating.median()
rating_fillvalue
```

```python
df.Rating.fillna(rating_fillvalue, inplace = True)
```

```python
df.isnull().sum()
```

```python
type_fillnul = df.Type.mode()
type_fillnul
```

```python
df.Type.fillna(type_fillnul[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
currentfillna = df['Current Ver'].mode()
currentfillna.values[0] # now it's working in the current ver column for filling na value by the values[0]
```

```python
df['Current Ver'].fillna(currentfillna.values[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
androidfillna = df['Android Ver'].mode().values[0]
androidfillna
```

```python
df['Android Ver'].fillna(androidfillna, inplace = True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.head()
```

```python
df.Price = df.Price.apply(lambda x: str(x).replace('$', "") if '$' in str(x) else str(x))
df.Price = df.Price.apply(lambda x: float(x))
```

```python
df.Price.mean()
```

```python
df.Reviews = pd.to_numeric(df.Reviews, errors='coerce')
```

```python
df.Installs = df.Installs.apply(lambda j: str(j).replace('+', "") if '+' in str(j) else str(j))

df.Installs = df.Installs.apply(lambda j : str(j).replace(',', "") if ',' in str(j) else str(j))
```

```python
df.Installs = df.Installs.apply(lambda j: float(j))
```
